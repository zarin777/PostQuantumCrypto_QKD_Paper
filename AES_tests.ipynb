```python
from google.colab import files
import os
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives import padding

# Function to encrypt a file using AES with a 256-bit key
def encrypt_file(file_path, key, iv):
    encrypted_data = b''

    # AES Cipher setup
    cipher = Cipher(algorithms.AES(key), modes.CBC(iv), backend=default_backend())
    encryptor = cipher.encryptor()

    # Padding setup
    padder = padding.PKCS7(128).padder()

    # Read the entire file content
    with open(file_path, 'rb') as file:
        data = file.read()

    # Apply padding
    padded_data = padder.update(data) + padder.finalize()

    # Encrypt the padded data
    encrypted_data = encryptor.update(padded_data) + encryptor.finalize()

    return data, padded_data, encrypted_data

# Function to process the file and print details
def process_file():
    # Upload file through Google Colab
    uploaded = files.upload()

    # Ensure a file was uploaded
    if not uploaded:
        print("No file uploaded.")
        return

    # Get file name and path
    file_name = next(iter(uploaded))
    file_path = f"./{file_name}"

    # Get user-defined 256-bit binary key
    binary_key = input("Enter a 256-bit binary key (256 bits = 32 bytes): ")
    if len(binary_key) != 256 or not all(bit in '01' for bit in binary_key):
        print("Invalid key. Please provide exactly 256 bits (0s and 1s).")
        return

    # Convert binary key to bytes
    key = int(binary_key, 2).to_bytes(32, byteorder='big')

    # Generate a random IV
    iv = os.urandom(16)  # AES block size IV (16 bytes)

    # Encrypt the file and retrieve data
    plaintext, padded_plaintext, ciphertext = encrypt_file(file_path, key, iv)

    # Convert to binary for display
    plaintext_binary = ''.join(format(byte, '08b') for byte in plaintext)
    padded_binary = ''.join(format(byte, '08b') for byte in padded_plaintext)
    ciphertext_binary = ''.join(format(byte, '08b') for byte in ciphertext)

    # Print details
    print("\nPlaintext (entire file in binary):\n", plaintext_binary)
    print("\nPadded Plaintext (in binary):\n", padded_binary)
    print("\nCiphertext (entire file in binary):\n",ciphertext_binary)

    # Print Key and IV for decryption
    print("\n256-bit Key (in binary, keep this safe):", binary_key)
    print("IV (in hex, keep this safe):", iv.hex())

# Main function to execute the process
process_file()
```

```python
from scipy.spatial.distance import hamming

# Function to calculate Hamming Distance using scipy
def calculate_hamming_distance():
    # User-provided plaintext and ciphertext in binary format
    plaintext_binary = input("Enter the plaintext in binary format: ")
    ciphertext_binary1 = input("Enter the ciphertext in binary format: ")

    # Adjust lengths if they are not the same
    min_length = min(len(plaintext_binary), len(ciphertext_binary))
    adjusted_plaintext = plaintext_binary[:min_length]
    adjusted_ciphertext = ciphertext_binary1[:min_length]

    # Calculate Hamming Distance for the aligned parts
    distance = hamming(list(adjusted_plaintext), list(adjusted_ciphertext)) * min_length

    print(f"Hamming Distance (aligned parts): {int(distance)}")
    print(f"Original lengths: Plaintext = {len(plaintext_binary)}, Ciphertext = {len(ciphertext_binary1)}")
    print(f"Compared lengths: {min_length}")
    percentage = (distance / min_length) * 100
    print(f"Hamming Distance Percentage: {percentage:.2f}%")



# Execute the function
calculate_hamming_distance()
```

```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad
from Crypto.Random import get_random_bytes

# Generate a random 128-bit AES key
key = get_random_bytes(16)

# User input for plaintext
plaintext = input("Enter the plaintext to encrypt: ")

# Encrypt the plaintext
cipher = AES.new(key, AES.MODE_ECB)
ciphertext = cipher.encrypt(pad(plaintext.encode('utf-8'), AES.block_size))
ciphertext_binary = ''.join(format(byte, '08b') for byte in ciphertext)
# Print the ciphertext in binary format
print("Ciphertext (binary):", ciphertext_binary)

```

```python
from collections import Counter

# Analyze frequency of bytes in ciphertext
def analyze_frequency(ciphertext):
    return Counter(ciphertext)

# User input for ciphertexts (binary strings)
ciphertext1 = input("Enter the first ciphertext (binary string): ")
ciphertext2 = input("Enter the second ciphertext (binary string): ")

# Convert binary strings to bytes
ciphertext1_bytes = bytes(int(ciphertext1[i:i+8], 2) for i in range(0, len(ciphertext1), 8))
ciphertext2_bytes = bytes(int(ciphertext2[i:i+8], 2) for i in range(0, len(ciphertext2), 8))

# Analyze frequencies
frequency1 = analyze_frequency(ciphertext1_bytes)
frequency2 = analyze_frequency(ciphertext2_bytes)

# Print results
print("\n--- Frequency Analysis ---")
print(f"Frequency Analysis for Ciphertext 1: {dict(frequency1)}\n")
print(f"Frequency Analysis for Ciphertext 2: {dict(frequency2)}")
```
